```python
mpl.rcParams.update(mpl.rcParamsDefault)

fig = plt.figure(figsize=(4, 3), dpi=100)
nRow = 1
nCol = 1
ax = plt.subplot(nRow, nCol, 1)

xDat = np.linspace(0,2*np.pi,100)
y1Dat = np.sin(xDat)
y2Dat = np.sin(2*xDat)
y3Dat = np.sin(3*xDat)
ax.plot(xDat, y1Dat)
ax.plot(xDat, y2Dat)
ax.plot(xDat, y3Dat)
        
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$f(x)$')
plt.savefig("sv1.pdf", bbox_inches='tight', format='pdf')
plt.show()
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)
plt.rc("font", family="Roboto Condensed")
plt.rc("xtick", labelsize="medium")
rc = {"mathtext.fontset" : "cm"}
plt.rcParams.update(rc)
plt.rcParams["font.serif"] = plt.rcParams["font.serif"]

linewidth = 1.2
linestyle = "-"
label_size = 10
color = '#EF6024'

fig = plt.figure(figsize=(5, 3), dpi=200)
nRow = 1
nCol = 1
ax = plt.subplot(nRow, nCol, 1)

xDat = np.linspace(0,2*np.pi,100)
y1Dat = np.sin(xDat)
y2Dat = np.sin(2*xDat)
y3Dat = np.sin(3*xDat)
ax.plot(xDat, y1Dat,
        color=colorsNw['red'][6],
        linewidth=linewidth, zorder=-1)
ax.plot(xDat, y2Dat,
        color=colorsNw['blue'][6],
        linewidth=linewidth, zorder=-2)
ax.plot(xDat, y3Dat,
        color=colorsNw['green'][6],
        linewidth=linewidth, zorder=-3)

ax.set_xlabel(r'$x$',usetex=True, fontsize=label_size+4)
ax.set_ylabel(r'$f(x)$', usetex=True, rotation=90, fontsize=label_size+4)
ax.tick_params(axis='both', which='major', labelsize=label_size)
ax.set_xticks([np.pi/2, np.pi, 1.5*np.pi, 2*np.pi])
ax.set_yticks([-1, -0.5, 0, 0.5, 1])
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1, 1)
labels = [r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$']
ax.set_xticklabels(labels, usetex=True)
ax.grid()

plt.tight_layout()
plt.savefig("sv2.pdf", bbox_inches='tight', format='pdf')
plt.show()
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)
font = {'family' : 'SF Pro',
        'weight' : 'light',
        'size'   : 22}

mpl.rc('font', **font)
plt.rc("xtick", labelsize="medium")
rc = {"mathtext.fontset" : "cm"}
plt.rcParams.update(rc)

linewidth = 1.0
linestyle = "-"
label_size = 10
color = '#EF6024'

fig = plt.figure(figsize=(4, 3), dpi=200)
nRow = 1
nCol = 1
ax = plt.subplot(nRow, nCol, 1)

xDat = np.linspace(0,2*np.pi,100)
y1Dat = np.sin(xDat)
y2Dat = np.sin(2*xDat)
y3Dat = np.sin(3*xDat)
ax.plot(xDat, y1Dat,
        color=colorsNw['indigo'][6],
        linewidth=linewidth, zorder=-1)
ax.plot(xDat, y1Dat,
        color='white',
        linewidth=linewidth+1, zorder=-2)
ax.plot(xDat, y2Dat,
        color=colorsNw['orange'][6],
        linewidth=linewidth, zorder=-3)
ax.plot(xDat, y2Dat,
        color='white',
        linewidth=linewidth+1, zorder=-4)
ax.plot(xDat, y3Dat,
        color=colorsNw['teal'][6],
        linewidth=linewidth, zorder=-5)
ax.plot(xDat, y3Dat,
        color='white',
        linewidth=linewidth+1, zorder=-6)

ax.set_xlabel(r'$x$',usetex=True, fontsize=label_size+4)
ax.set_ylabel(r'$f(x)$', usetex=True, rotation=90, fontsize=label_size+4)
ax.tick_params(axis='both', which='major', labelsize=label_size)
ax.set_yticks([-1, -0.5, 0.5, 1])
ax.set_xticks([np.pi/2, np.pi, 1.5*np.pi, 2*np.pi])
labels = [r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$']
ax.set_xticklabels(labels, usetex=True)

ax.xaxis.set_label_coords(0.5,-0.1)
ax.yaxis.set_label_coords(-0.1,0.45)

ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["left"].set_position(("data", 0))
ax.spines["bottom"].set_position(("data", 0))

ax.set_aspect(1.4)
plt.tight_layout()
plt.savefig("sv3.pdf", bbox_inches='tight', format='pdf')
plt.show()
```
